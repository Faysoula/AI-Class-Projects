```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Reading the data
data = pd.read_csv('/content/spotify-2023.csv', encoding='ISO-8859-1')

# Check for missing values in bpm and danceability_%
print(data[['bpm', 'danceability_%']].isnull().sum())

# Replace NaN with a suitable value or drop rows if necessary
data['bpm'].fillna(0, inplace=True)  # replace NaN with 0
data['danceability_%'].fillna(data['danceability_%'].mean(), inplace=True)  # Replace NaN with mean

# Separate features (X) and target (y)
X = data[['bpm']].values
y = data['danceability_%'].values

# Create a LinearRegression model
model = LinearRegression()

# Fit the model to the data
model.fit(X, y)

# Predict using the model
X_new = np.array([[120]])  # Example BPM
prediction = model.predict(X_new)

# Print the predicted danceability percentage for a given BPM value
print("Predicted Danceability_% for BPM :", prediction[0])

# Set the size of the plot
plt.figure(figsize=(8, 4))

# Create a scatter plot with BPM on the X-axis and danceability percentage on the Y-axis
plt.scatter(X, y)

# Label the X-axis as "BPM"
plt.xlabel("BPM")

# Label the Y-axis as "Danceability %"
plt.ylabel("Danceability %")

# Give the plot a title "BPM vs. Danceability %"
plt.title("BPM vs. Danceability %")

# Display the plot
plt.show()
```

The problem here is trying to predict the dancebility of songs on spotify, which we are solving by training our model using a dataset of the top rated songs on spotify.
